```python

import pandas as pd
# to get math structures
import numpy as np
# library to draw
import matplotlib.pyplot as plt
```

# Comparison of stationary models

## Reading data

```python
arch_mo ="./data/stationary_mo_crra_0.5_100_step200.csv"


db_mo = pd.read_csv(arch_mo)

db_mo.head()
```

```python
arch_st = "./data/stationary_standard_crra_0.5_100_step200.csv"

db_st = pd.read_csv(arch_st)

db_st.head()
```

```python
# colors
cool_red = '#fe4a49'
cool_blue = '#2ab7ca'
```

## Plots

```python
plt.figure(figsize= (6,5))

plt.plot(db_mo["u_princ"], db_mo["u_agent"],
            color= cool_blue,
            alpha=1,
            label = "mo bargaining")

plt.plot(db_st["u_princ"], db_st["u_agent"],
            color= cool_red,
            alpha=1,
            label = "standard")

plt.legend( markerscale=3.3, scatterpoints=1)#, fontsize=10)
plt.xlabel("Principal Expected Utility")
plt.ylabel("Agent Expected Utility")

plt.suptitle('Pareto Frontier', fontsize=10)
plt.title('Stationary, h=0.50',fontsize=8)


plt.savefig('./reports/Stationary_comparison_frontier'+'.pdf')
```

```python
plt.figure(figsize= (6,5))

plt.scatter( db_mo["u_agent"], db_mo["w_h"],
            color= cool_blue,
            marker='^',
            s = 20.0,
            alpha=1,
           label = "high wage mo bargaining")


plt.scatter( db_st["u_agent"], db_st["w_h"],
            color= cool_red,
            marker='^',
             s = 20.0,
            alpha=1,
            label = "high wage standard")

plt.scatter( db_mo["u_agent"], db_mo["w_l"],
            color= "darkblue",
            marker='v',
            s = 20.0,
            alpha=1,
           label = "low wage mo bargaining")


plt.scatter( db_st["u_agent"], db_st["w_l"],
            color= "darkred",
            marker='v',
             s = 20.0,
            alpha=1,
            label = "low wage standard")

plt.legend( markerscale=1.5, scatterpoints=1)#, fontsize=10)
plt.xlabel("agent utility")
plt.ylabel("wage")

plt.suptitle('Wages vs Agent Utility', fontsize=10)
plt.title('Stationary, h=0.50',fontsize=8)

plt.savefig('./reports/Stationary_comparison_wages_utility'+'.pdf')
```

```python
plt.figure(figsize= (10,5))


plt.subplot(121)
plt.scatter( db_mo["u_agent"], db_mo["w_h"],
            color= cool_blue,
            marker='^',
            s = 20.0,
            alpha=1,
           label = "high wage mo bargaining")


plt.scatter( db_st["u_agent"], db_st["w_h"],
            color= cool_red,
            marker='^',
             s = 20.0,
            alpha=1,
            label = "high wage standard")

plt.legend( markerscale=1.5, scatterpoints=1)#, fontsize=10)
plt.xlabel("agent utility")
plt.ylabel("high wage")
plt.title('Stationary, h=0.50',fontsize=8)

plt.subplot(122)
plt.scatter( db_mo["u_agent"], db_mo["w_l"],
            color= "darkblue",
            marker='v',
            s = 20.0,
            alpha=1,
           label = "low wage mo bargaining")


plt.scatter( db_st["u_agent"], db_st["w_l"],
            color= "darkred",
            marker='v',
             s = 20.0,
            alpha=1,
            label = "low wage standard")

plt.legend( markerscale=1.5, scatterpoints=1)#, fontsize=10)
plt.xlabel("agent utility")
plt.ylabel("low wage")

plt.suptitle('Wages vs Agent Utility', fontsize=10)
plt.title('Stationary, h=0.50',fontsize=8)

plt.savefig('./reports/Stationary_comparison_wages_utility_1x2'+'.pdf')
```

```python
plt.figure(figsize= (6,5))

plt.scatter(db_mo["w_l"], db_mo["w_h"],
            color= cool_blue,
            alpha=1,
           label = "mo bargaining")


plt.scatter(db_st["w_l"], db_st["w_h"],
            color= cool_red,
            alpha=1,
            marker= 'p',
            label = "standard")

plt.legend( markerscale=1.5, scatterpoints=1)#, fontsize=10)
plt.xlabel("low wage")
plt.ylabel("high wage")

plt.suptitle('Wages', fontsize=10)
plt.title('Stationary, h=0.50',fontsize=8)

plt.savefig('./reports/Stationary_comparison_wages'+'.pdf')
```

```python
plt.figure(figsize= (6,5))

ul_mo = [ db_mo["u_agent"][i-1] for i in db_mo["u_l"]]
uh_mo= [ db_mo["u_agent"][i-1] for i in db_mo["u_h"]]

ul_st = [ db_st["u_agent"][i-1] for i in db_st["u_l"]]
uh_st = [ db_st["u_agent"][i-1] for i in db_st["u_h"]]


plt.scatter(ul_st, ul_st,
            s= 8.0,
            color= cool_red,
            alpha=1,
            marker = "p",
            label = "standard")


plt.scatter(ul_mo, uh_mo,
            color= cool_blue,
            s=8.0,
            alpha=1,
            marker = "o",
            label = "mo bargaining")


plt.legend( markerscale=1.5, scatterpoints=1)#, fontsize=10)
plt.xlabel("Low Future compensation")
plt.ylabel("High Future compensation")

plt.suptitle('Future Compensations', fontsize=10)
plt.title('Stationary, h=0.50',fontsize=8)


plt.savefig('./reports/Stationary_comparison_future'+'.pdf')
```

```python
plt.figure(figsize= (6,5))

plt.scatter( db_mo["u_agent"],db_mo["a"],
            color= cool_blue,
            alpha=1,
            label = "mo bargaining")


plt.scatter(db_st["u_agent"],db_st["a"], 
            color= cool_red,
            alpha=1,
            marker = "p",
            label = "standard")

plt.legend( markerscale=1.5, scatterpoints=1)#, fontsize=10)
plt.xlabel("Expected utility agent")
plt.ylabel("effort")

plt.suptitle('Efforts', fontsize=10)
plt.title('Stationary, h=0.50',fontsize=8)

plt.savefig('./reports/Stationary_comparison_effort'+'.pdf')
```
